```python
song_grouped = song_df.groupby(['ARTIST_NAME']).agg({'PLAYS':'count'}).reset_index()
song_grouped.head()

```

```python
grouped_sum = song_grouped['PLAYS'].sum()
grouped_sum
```

```python
song_grouped['PERCENTAGE'] = (song_grouped['PLAYS'] / grouped_sum ) * 100
song_grouped.sort_values(['PLAYS', 'ARTIST_NAME'], ascending=[0,1])
```
